```python
# 무역수지 컬럼 추가 작업
data_무역수지 = pd.read_csv('data/수출입총괄.csv')
data_무역수지.drop(labels=['수출건수', '수출금액', '수입건수', '수입금액', '(단위1000USD)'], axis=1, inplace=True)
data_무역수지1 = data_무역수지[1:]
data_무역수지1
```

```python
# 데이터 병합전 전처리 작업 (날짜 형식 & 무역수지 값 형식(쉼표 제거))
date_li = []
for date in data_무역수지1['기간']:
    date_li.append(date.replace('.','-'))

eco_li = [] # 무역수지 empty list
for eco in data_무역수지1['무역수지']:
    eco_li.append(eco.replace(',',''))

data = {
    '날짜':date_li,
    '무역수지':eco_li
}

df_무역수지 = pd.DataFrame(data)
df_무역수지
```

```python
# 10월 데이터 값이 1로 표현되어 전처리 작업 수행

# 1. '-' 제거 작업
oct_li = []
for d in df_무역수지['날짜']:
    oct_li.append(d.split('-'))

# 2. 10월달 해당 값이 1로 표현됨 : 1 -> 10 변경 전처리 작업
for i in range (len(oct_li)):
    if len(oct_li[i][1]) == 1:
        oct_li[i][1] += '0'
oct_li

# 3. 20xx-xx 형식으로 다시 concate 후 oct_li1로 append 
oct_li1 = []
for date in oct_li:
    d = date[0] + '-' + date[1]
    oct_li1.append(d)
oct_li1
```

```python
data1 = {
    '날짜':oct_li1,
    '무역수지':eco_li
}

df_무역수지1 = pd.DataFrame(data1)
df_무역수지1[14:]
```

```python
# 1803 2207 기간 설정
df_무역수지2 = df_무역수지1[14:]
df_무역수지2.index = (np.arange(1, len(df_무역수지2)+1))
df_무역수지2
```
